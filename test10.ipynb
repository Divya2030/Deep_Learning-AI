```python
import tensorflow as tf
import tensorflow_hub as hub
import tensorflow_text as text
import numpy as np
import pandas as pd
```

##Getting to know about BERT

```python
preprocess_url='https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3'

model_url='https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4'
```

```python
bert_preprocess_url=hub.KerasLayer(preprocess_url)
```

```python
analyze_text=['I like python programming','This is a deep learning class','All my colleagues like deep learning']
```

```python
preprocess_text = bert_preprocess_url(analyze_text)
```

```python
bert_model_url=hub.KerasLayer(model_url)
```

```python
bert_model_results=bert_model_url(preprocess_text)
```
